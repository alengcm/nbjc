```python
import os

def rename_images(folder_path):
    # Supported image extensions
    valid_extensions = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.webp')
    
    # Get a list of all files in the directory
    files = [f for f in os.listdir(folder_path) if f.lower().endswith(valid_extensions)]
    
    # Sort files to ensure they are renamed in order
    files.sort()

    count = 1
    for filename in files:
        # Construct the new name
        extension = os.path.splitext(filename)[1]
        new_name = f"{count}{extension}"
        
        # Define full file paths
        source = os.path.join(folder_path, filename)
        destination = os.path.join(folder_path, new_name)
        
        # Rename the file
        try:
            os.rename(source, destination)
            print(f'Renamed: {filename} -> {new_name}')
            count += 1
        except FileExistsError:
            print(f"Skipping {filename}: {new_name} already exists.")

# --- Settings ---
target_folder = 'C:/Users/alen/Documents/nbjc/nbjc/gallery' # Update this path!
rename_images(target_folder)
```
